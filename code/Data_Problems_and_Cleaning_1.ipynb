# Làm sạch lần 1
Loại bỏ giá trị NaN và thay thế giá trị

## Thay thế giá trị trước khi drop

Không thay thế giá trị NaN ở cột ``Num of MC/SC`` thành 0 bởi cũng có thể có nguyên nhân web chưa có dữ liệu. 

Vì vậy chỉ thay thế bản ghi có giá trị ``Num of MC/SC`` là 1 và giá trị NaN ở ``Pic MC/SC``

```python
dataNew = data.copy()
```

### Camera sau (MC)

```python
VGA = '0.3'
```

```python
idx_VGA = data_VGA[(data_VGA['Num of MC'] == '1.0')].index
```

```python
for i in idx_VGA:
    dataNew.at[i,'Pic MC'] = VGA
```

```python
dataNew.loc[[994]]
```

Tại index 994, cần thay thành 13, 5 ở Pic MC, 1080p ở Video MC, 5MP ở Pic SC

```python
dataNew.at[994,'Pic MC'] = '13,5'
dataNew.at[994,'Video MC'] = '1080p@30fps'
dataNew.at[994,'Pic SC'] = '5'
```

### Camera trước (SC)
Một số máy có 2 màn hình sẽ chỉ có 1 cụm camera đóng vai trò làm cả Camera trước và sau. 

Vì vậy thay giá trị ``Pic SC`` giống với ``Pic MC``, ``Video SC`` giống với ``Video MC``

```python
lst_MC_SC
```

```python
for i in lst_MC_SC:
    dataNew.at[i,'Pic SC'] = dataNew.at[i,'Pic MC']
    dataNew.at[i,'Video SC'] = dataNew.at[i,'Video MC'] 
```

Các máy có chất lượng ảnh VGA (0.3MP)

```python
idx_VGA_SC = data_PicSC[data_PicSC['Num of SC'] == '1.0'].index
idx_VGA_SC
```

```python
for i in idx_VGA_SC:
    dataNew.at[i,'Pic SC'] = VGA
```

## Drop cột Video SC và giá trị NaN

```python
df = dataNew.drop(columns=['Video SC']).dropna()
```

```python
df.shape
```

Vậy tập dữ liệu mới đã loại bỏ 1 cột và còn 4884 dữ liệu không bị NaN
